```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# count missing values in training data
df = pd.read_csv('training_data.csv')
missing_values = df.isnull().sum()
print('Missing values in training data:', missing_values)

# total missing values in training data
total_missing_values = missing_values.sum()
print('Total missing values in training data:', total_missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset without header
data = pd.read_csv("training_data_targets.csv", header=None, names=['target'])

# Assuming the target variable is in the first column
class_distribution = data.iloc[:, 0].value_counts()

# Plotting the class distribution curve with thinner bars
plt.figure(figsize=(4, 5))
sns.barplot(x=class_distribution.index, y=class_distribution.values, palette="viridis", saturation=0.75)
plt.title('Class Distribution in the Dataset')
plt.xlabel('Class')
plt.ylabel('Number of Instances')
plt.show()
```
